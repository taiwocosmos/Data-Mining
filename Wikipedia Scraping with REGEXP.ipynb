```python
import requests
from bs4 import BeautifulSoup
```

```python
base = 'https://en.wikipedia.org/wiki/Main_Page'
result = requests.get(base)
result
```

```python
content = result.text
```

```python
soup = BeautifulSoup(content, 'lxml')
print(soup.prettify())
```

```python
contents = soup.find('div', {'id': 'mp-upper'})
```

```python
import re
links = contents.findAll('a', href=re.compile(r"(/wiki/)+([A-Za-z0-9_:()]+)"))
len(links)
```

```python
title = []
all_links = []
for link in links:
    if len(link.text)== 0:
        title.append('No Title')
    else:
        title.append(link.text)
    all_links.append(link['href'])
all_links
```

```python
import pandas as pd
n = range(1, (len(all_links)+1))
ind = pd.Index(n)
dataset = pd.DataFrame({'Title': title,
                       'Link': all_links}, index=ind)
dataset
```
