```python
from selenium import webdriver 
import chromedriver_autoinstaller

chromedriver_autoinstaller.install()
```

```python
website = 'https://www.adamchoi.co.uk/'

driver = webdriver.Chrome()

driver.get(website)
```

```python
from selenium.webdriver.common.by import By
loc = '//table[@id= "desktop-streaks-table"]'
one_5 = driver.find_element(By.XPATH, loc)
one_5
```

```python
loc_ = '//tbody[@class= "ng-scope"]'
t_row = driver.find_elements(By.XPATH, loc_)
len(t_row)
```

```python
import numpy as np
import pandas as pd

tds1 = []
tds2 = []
tds3 = []
tds4 = []

for row in t_row:
    td1 = row.find_element(By.XPATH, './tr/td[2]')
    if len(td1.text)==0:
        tds1.append(np.nan)
    else:
        tds1.append(td1.text)
    td2 = row.find_element(By.XPATH, './tr/td[3]')
    if len(td2.text)==0:
        tds2.append(np.nan)
    else:
        tds2.append(td2.text)
    td3 = row.find_element(By.XPATH, './tr/td[4]')
    if len(td3.text)==0:
        tds3.append(np.nan)
    else:
        tds3.append(td3.text)
        
# tds3
```

```python
dataset = pd.DataFrame({'League': tds1, 'Stat': tds2, 'Next Match': tds3})
dataset
```

```python
dataset.dropna(inplace=True)
```

```python
dataset
```

```python
fixtures = []
for i in range (1, 28):
    fixtures.append('Match ' +str(i))
fixtures
```

```python
ind = pd.Index(fixtures)
dataset.set_index(ind)
```

```python
dataset.to_csv('AdamChoi.scraped.csv')
```
